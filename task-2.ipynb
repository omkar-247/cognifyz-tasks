```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
df.head()
```

Task: Votes Analysis

Identify the restaurants with the highest and
lowest number of votes.

```python
highest_voted= df.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending = False)
print('Highest voted restaurant:',highest_voted.idxmax())
```

```python
lowest_voted= df.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending = True)
print('lowest voted restaurant:',lowest_voted.idxmin())
```

Analyze if there is a correlation between the
number of votes and the rating of a
restaurant.

```python
corelation = np.corrcoef(df['Votes'],df['Aggregate rating'])
corelation = np.round(corelation , 2)
corelation
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.heatmap(corelation,annot = True ,fmt='.2f',linewidth= .4)
plt.show()
```

```python
sns.scatterplot(data = df,y = 'Votes',hue='Aggregate rating',x = 'Aggregate rating')
sns.regplot(data = df,y = 'Votes',x = 'Aggregate rating',scatter = False)
plt.title(f'correlation between the Votesand the Aggregate rating:0.31')
plt.show()
```
