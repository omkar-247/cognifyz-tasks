```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
```

```python
df.columns
```

Task: Price Range vs. Online Delivery and
Table Booking

Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.

```python
x1 = df['Has Online delivery']
y = df['Average Cost for two']
x2 = df['Has Table booking']
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x1,y=y)
plt.xlabel('Online delivery')
plt.ylabel('Average Cost for two')
plt.title('Online delivery effected by cost')
plt.subplot(1,2,2)
sns.boxplot(x=x2,y=y)
plt.xlabel('Table booking')
plt.ylabel('Average Cost for two')
plt.title('Table booking effected by cost')
plt.tight_layout()
plt.show()
```

```python
print('Both online delivery and Table booking are negetively impacted by the cost')
```

Determine if higher-priced restaurants are
more likely to offer these services.

lets checkout for the restaurants have average cost less than 10,000

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x1,y='Average Cost for two', data = df.loc[df['Average Cost for two']<10000])
plt.title('Online delivery by cost')
plt.subplot(1,2,2)
sns.boxplot(x=x2,y='Average Cost for two', data = df.loc[df['Average Cost for two']<10000])
plt.title('Online delivery by cost')
plt.tight_layout()
plt.show()
```

```python
print('The Online Delivery facility is only available for those restaurant, where the avg cost is less than 1000')
print('The Table booking facility is avaliable for restaurants , where the avg cost is in range from 1000 to 2000. ')
```
